```python
import matplotlib.pyplot as plt
import cv2
import statistics
import numpy as np
```

```python
def escalaGrisesPonderada(imagen):
    image = cv2.imread(imagen)
    ancho = image.shape[0]
    alto = image.shape[1]
    gray = np.zeros((ancho,alto))
    for i in range(0,ancho):
        for j in range(0,alto):
            pixel = image[i,j]
            gray[i,j]=( (0.3 * pixel[2]) + (0.59 * pixel[1]) + (0.11 * pixel[0]) )
    cv2.imwrite('imagenponderado.jpg', gray) 
    
    imagenponderada = cv2.imread('imagenponderado.jpg')
    imagenponderada = cv2.cvtColor(imagenponderada, cv2.COLOR_BGR2RGB)
    plt.imshow(imagenponderada)
    plt.show()
escalaGrisesPonderada('../imagen1.jpg')
```
