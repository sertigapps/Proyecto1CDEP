# Ahora vamos a cargar los modelos

```python
decissionTree = pickle.load(open('DecissionTree.sav', 'rb'))
SVM = pickle.load(open('SVM.sav', 'rb'))
bayes = pickle.load(open('bayes.pkl', 'rb'))
def probInd(X, ant, dists):
    prod = np.prod(np.array([d.pdf(z) for (d, z) in zip(dists, X)]))
    return ant * prod
def predictBayes(x):
    p0 = probInd(x, bayes['pA0'], bayes['nY0'])
    p1 = probInd(x, bayes['pA1'], bayes['nY1'])
    if p0>p1:
        return 0
    else:
        return 1
def predictoreBayes(x):
    predictor = lambda t: predictBayes(t)
    predicciones =  np.array([predictor(xi) for xi in x])
    return predicciones
modeloreglog=pd.read_csv('modeloreg.csv',header=None)
def predictLog(x):
    logitspred = np.matmul(x,np.array(modeloreglog))
    predict = np.argmax(logitspred, axis=1)
    return predict
def funcionPrueba(x,y):
    dt = decissionTree.predict(x)[0]
    svm = predictLog(x)[0]
    bayes = predictoreBayes(x)[0]
    log = predictLog(x)[0]
    essemble = 0
    if ( dt + svm + bayes + log == 2 ):
        essemble = dt
    if ( dt + svm + bayes + log > 2 ):
        essemble = 1
    print('Prediccion Essemble ' + str(essemble))
    print('Prediccion SVM ' + str(svm))
    print('Prediccion bayes ' + str(bayes))
    print('Prediccion DT ' + str(dt))
    print(' Valor Esperado era ' + str(y) + ' Valor de Essemble fue ' +str(essemble))
    
    plt.figure(figsize=(50,50))
    a = plot_tree(decissionTree, 
                  filled=True, 
                  rounded=True, 
                  fontsize=14)

x_train, x_test10, y_train, y_test0 = train_test_split(
     x_test, y_test, test_size=10, shuffle=True)
x_test10
    
```
