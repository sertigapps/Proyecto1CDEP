```python
import matplotlib.pyplot as plt
import cv2
import statistics
import numpy as np
```

```python
def generarBordesyContornos(img):
    img = cv2.imread(img,0)
    edges = cv2.Canny(img,100,200)
    plt.imshow(img,cmap = 'gray')
    plt.title('Imagen Original'), plt.xticks([]), plt.yticks([])
    plt.show()
    plt.imshow(edges,cmap = 'gray')
    plt.title('Bordes Image'), plt.xticks([]), plt.yticks([])
    
    plt.show()
    
    ret, thresh = cv2.threshold(img, 127, 255, 0)
    im2, contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    plt.imshow(im2,cmap = 'gray')

    plt.title('Contornos Image'), plt.xticks([]), plt.yticks([])
    
    plt.show()

generarBordesyContornos('../carvana.jpg')
```

```python

generarBordesyContornos('../theo.jpg')
```

# Descripcion
Los bordes se calculan como puntos que son extremos del gradiente de la imagen en la dirección del gradiente. Por ejemplo, podemos pensar en ellos como los puntos mínimo y máximo en una función en una dimension. El concepto es que los píxeles de borde son una minimo o maximo local: solo señalan una diferencia significativa entre los píxeles vecinos o cercanos.

Los contornos a menudo se obtienen de los bordes, pero están destinados a ser contornos de objetos. Por lo tanto, deben ser curvas cerradas. Puede pensar en ellos como fronteras (algunos algoritmos y bibliotecas de procesamiento de imágenes los llaman así). Cuando se obtienen de los bordes, se debe conectar los bordes para obtener un contorno cerrado.
