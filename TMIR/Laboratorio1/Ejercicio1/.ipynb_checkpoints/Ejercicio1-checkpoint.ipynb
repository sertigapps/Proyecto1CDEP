```python
import matplotlib.pyplot as plt
import cv2
import statistics
import numpy as np
```

# Ejercicio #1

```python
def generarBordesyContornos(img):
    img = cv2.imread(img,0)
    edges = cv2.Canny(img,100,200)
    plt.subplot(121),plt.imshow(img,cmap = 'gray')
    plt.title('Original Image'), plt.xticks([]), plt.yticks([])
    plt.subplot(122),plt.imshow(edges,cmap = 'gray')
    plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
    plt.show()
    
    edges = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=9)
    plt.subplot(121),plt.imshow(img,cmap = 'gray')
    plt.title('Original Image'), plt.xticks([]), plt.yticks([])
    plt.subplot(122),plt.imshow(edges,cmap = 'gray')
    plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
    plt.show()
    
    edges = cv2.Laplacian(img,cv2.CV_64F)
    plt.subplot(121),plt.imshow(img,cmap = 'gray')
    plt.title('Original Image'), plt.xticks([]), plt.yticks([])
    plt.subplot(122),plt.imshow(edges,cmap = 'gray')
    plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
    plt.show()
    
    
    
generarBordes('../LogoRM.jpeg')
```

# Desventajas
| Canny | Sobel | Laplace|
|-------|-------|--------|
|La principal desventaja de usar el detector de bordes Canny es que consume mucho tiempo debido a su compleja computacin.|Solo detecta bordes gruesos en su mayoria|  Mucho mas sensitivo a ruido
|Es difcil de implementar para alcanzar una respuesta en tiempo real.|No se comporta bien con imagenes ruidosas | Los bordes claros se ven afectados por el ruido
|Da un sesgo hacia los bordes verticales y horizontales y no da una buena aproximacin de la simetra rotacional.|
