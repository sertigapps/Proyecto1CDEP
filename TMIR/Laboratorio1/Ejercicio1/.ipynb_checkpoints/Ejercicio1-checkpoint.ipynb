```python
import matplotlib.pyplot as plt
import cv2
import statistics
import numpy as np
```

# Ejercicio #1

```python
def generarBordesyContornos(img):
    img = cv2.imread(img,0)
    edges = cv2.Canny(img,100,200)
    plt.subplot(121),plt.imshow(img,cmap = 'gray')
    plt.title('Original Image'), plt.xticks([]), plt.yticks([])
    plt.subplot(122),plt.imshow(edges,cmap = 'gray')
    plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
    plt.show()
    
    edges = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=9)
    plt.subplot(121),plt.imshow(img,cmap = 'gray')
    plt.title('Original Image'), plt.xticks([]), plt.yticks([])
    plt.subplot(122),plt.imshow(edges,cmap = 'gray')
    plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
    plt.show()
    
    edges = cv2.Laplacian(img,cv2.CV_64F)
    plt.subplot(121),plt.imshow(img,cmap = 'gray')
    plt.title('Original Image'), plt.xticks([]), plt.yticks([])
    plt.subplot(122),plt.imshow(edges,cmap = 'gray')
    plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
    plt.show()
    
    
    
generarBordes('../LogoRM.jpeg')
```

# Ventajas
| Canny | Sobel | Laplace|
|-------|-------|--------|
|La presencia del filtro gaussiano permite eliminar cualquier ruido en una imagen.| Su simplicidad porque aproxima los gradientes| Es aun mas rapido que sobel ya que usa solamente un kernel
|La señal se puede mejorar con respecto a la relación de ruido mediante el método de supresión que da como resultado puntas de un píxel de ancho como salida.| Debido a su simplicidad su velocidad es rapida| Requiere menos operaciones
|Detecta los bordes en un estado ruidoso aplicando el método de umbral.| | Como el laplaciano se calcula de antemano solo una convolucion se aplica a la imagen
|La efectividad se puede ajustar mediante el uso de parámetros.|
|Da una buena localización, respuesta y es inmune a un ambiente ruidoso.|


# Desventajas
| Canny | Sobel | Laplace|
|-------|-------|--------|
|La principal desventaja de usar el detector de bordes Canny es que consume mucho tiempo debido a su compleja computación.|Solo detecta bordes gruesos en su mayoria|  Mucho mas sensitivo a ruido
|Es difícil de implementar para alcanzar una respuesta en tiempo real.|No se comporta bien con imagenes ruidosas | Los bordes claros se ven afectados por el ruido
|Da un sesgo hacia los bordes verticales y horizontales y no da una buena aproximación de la simetría rotacional.|
