```python
import matplotlib.pyplot as plt
import cv2
```

```python
def crearEscala3D(imagen):
    print('Imagen Original')
    image = cv2.imread(imagen)
    imageor = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.axis("off")
    plt.imshow(imageor)
    plt.show()
    copiaazul = image.copy()
    copiaverde = image.copy()
    copiarojo = image.copy()
    
    copiaazul[:, :, 1] = 0
    copiaazul[:, :, 2] = 0
    
    copiaazulgray = cv2.cvtColor(copiaazul, cv2.COLOR_BGR2GRAY)
    cv2.imwrite('ej3_salida_gray_azul.jpg', copiaazulgray) 
    
    copiaverde[:, :, 0] = 0
    copiaverde[:, :, 2] = 0
    
    copiaverdegray = cv2.cvtColor(copiaverde, cv2.COLOR_BGR2GRAY)
    cv2.imwrite('ej3_salida_gray_verde.jpg', copiaverdegray) 
    
    copiarojo[:, :, 0] = 0
    copiarojo[:, :, 1] = 0
    
    copiarojogray = cv2.cvtColor(copiarojo, cv2.COLOR_BGR2GRAY)
    cv2.imwrite('ej3_salida_gray_rojo.jpg', copiarojogray) 
crearEscala3D('../imagen1.jpg')
```
