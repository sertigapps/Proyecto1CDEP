Los teoremas matemáticos a menudo pueden generalizar desde dimensiones bajas a dimensiones superiores. Si sabemos cómo calcular gradientes para una función con dos variables, entonces probablemente también podremos descubrir cómo hacerlo para funciones con múltiples variables. Sin embargo, este no es el caso cuando intenta generalizar su estimador de densidad bidimensional a cientos o miles de dimensiones. Teóricamente, su estimador aún debería funcionar en alta dimensión. Pero eso requiere que recopilemos una cantidad exponencial de datos de entrenamiento para evitar los errores. En la práctica, los datos de entrenamiento disponibles a menudo son bastante limitados.

Un estimador bidimensional deja de funcionar en alta dimensión porque los datos de entrenamiento limitados en su mano no son estadísticamente significativos debido a su esparcimiento, es decir, no hay suficiente información para que pueda recuperar la verdadera distribución global. El concepto de significancia estadística puede ilustrarse en el caso unidimensional de la siguiente manera. Suponga que está extrayendo algunos puntos de una distribución normal e intentando estimar sus parámetros. Hace una gran diferencia si dibuja solo 10 puntos y si saca 100 puntos de la distribución.


![1 Dimension](img5.png)
![1 Dimension](img4.png)
![1 Dimension](img6.png)

En lo anterior, el primero es la verdadera distribución normal subyacente. El segundo se estima a partir de 10 puntos. El tercero se estima a partir de 100 puntos. Pocos puntos de entrenamiento resultan en una estimación pobre.

La maldición de la dimensionalidad describe un escenario en el que no es posible obtener un número exponencial de puntos de entrenamiento en un espacio de altas dimensiones, en el ejmplo anterior imaginemonos que no se pueden obtener fácilmente más de 10 puntos de entrenamiento, lo que resulta en una gran dificultad para estimar de manera confiable la función de densidad. Esta dificultad también surge al buscar un límite de decisión entre clases usando un clasificador. Nunca se sabe si el límite trazado para separar los puntos de entrenamiento apenas separa algunas muestras de entrenamiento especiales de toda la población o si realmente se generaliza bien a otros datos. La falta de capacidad para estimar y generalizar de manera confiable se llama sobreajuste. Incluso con suficientes datos de entrenamiento en un espacio de baja dimensión, el sobreajuste puede existir debido a varias razones (por ejemplo, alta complejidad del modelo, recopilación incorrecta de datos de entrenamiento, etc.). Sin embargo, en un espacio de alta dimensión, es más probable que ocurra un sobreajuste debido a la falta de datos de entrenamiento.

Soluciones para la maldicion de dimensionalidad:

 - Extraccion de Features
 - Reduccion de Dimensionalidad
 - Kernels
# Ejemplo Practico

Vamos a crear una funcion que reciba una serie de variables y un grado y con esto vamos a calcular la ecuacion resultante

```python
import numpy as np 
def obtenerEcuacion(variables, grado):
    varEcu = []
    varRes = []
    for i in range(len(variables)):
        for j in range(grado):
            if j == 0 :
                varEcu.append(variables[i])
            else:
                varEcu.append(variables[i]+ '^' + str(j+1))
    for i in range(len(varEcu)):
        varRes.append(varEcu[i])
        for j in range(len(varEcu)):
            if varEcu[i][0] != varEcu[j][0]:
                if varEcu[j]+'*'+varEcu[i] not in varRes:
                    varRes.append(varEcu[i]+'*'+varEcu[j])
    print(' + '.join(varRes))
```

Imaginemonos que tenemos 4 variables de grado 1 tendriamos las siguiente ecuacion


```python

obtenerEcuacion(np.array(['X','Y','Z','M']),1)
```

Imaginemonos que tenemos 4 variables de grado 5 tendriamos las siguiente ecuacion

```python

obtenerEcuacion(np.array(['X','Y','Z','M']),5)
```

Podemos ver como se aumenta la ecuacion en una forma muy grande ahora veamos 10 variables de grado 10

```python

obtenerEcuacion(np.array(['X','Y','Z','M','A','B','C','D','E','F']),10)
```

Podemos ver como se salio de las manos de forma rapida y sin escala
