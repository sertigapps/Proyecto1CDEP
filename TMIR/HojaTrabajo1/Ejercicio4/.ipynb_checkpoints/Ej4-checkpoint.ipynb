```python
import matplotlib.pyplot as plt
import cv2
import statistics
import numpy as np
```

```python
def mostrarHistogramas(imagen):
    image = cv2.imread(imagen)
    for i, col in enumerate(['b', 'g', 'r']):
        hist = cv2.calcHist([image], [i], None, [256], [0, 256])
        plt.plot(hist, color = col)
        plt.xlim([0, 256])
        plt.axvline(statistics.median(image[:, :, i].flatten()), 0, 30000, label='Median')
        plt.show()
    ancho = image.shape[0]
    alto = image.shape[1]
    gray = np.zeros((ancho,alto,1))
    for i in range(0,ancho):
        for j in range(0,alto):
            pixel = image[i,j]
            gray[i,j]=statistics.median(pixel)
    a = gray.flatten()
    plt.hist(a,histtype="step", bins = range(256), color = 'black') 
    plt.axvline(statistics.median(a), 0, 30000, label='Median')
    plt.show()
mostrarHistogramas('../imagen1.jpg')
```
