```python
import matplotlib.pyplot as plt
import cv2
import statistics
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
def transfInversa(im):
    image = cv2.imread(im)
    ancho = image.shape[0]
    alto = image.shape[1]
    gray = np.zeros((ancho,alto))
    for i in range(0,ancho):
        for j in range(0,alto):
            pixel = image[i,j]
            gray[i,j]=255 - ( (0.3 * pixel[2]) + (0.59 * pixel[1]) + (0.11 * pixel[0]) )
    cv2.imwrite('binary.jpg', gray) 
    
    imagenponderada = cv2.imread('binary.jpg')
    imagenponderada = cv2.cvtColor(imagenponderada, cv2.COLOR_BGR2RGB)
    plt.imshow(imagenponderada)
    plt.show()
transfInversa('../LogoRM2.jpeg')
```

```python
def transfLineal(im, m ,b):
    image = cv2.imread(im)
    ancho = image.shape[0]
    alto = image.shape[1]
    gray = np.zeros((ancho,alto))
    for i in range(0,ancho):
        for j in range(0,alto):
            pixel = image[i,j]
            gray[i,j]=( (0.3 * pixel[2]) + (0.59 * pixel[1]) + (0.11 * pixel[0]) )
            gray[i,j]=m * gray[i,j] + b
    cv2.imwrite('binary.jpg', gray) 
    
    imagenponderada = cv2.imread('binary.jpg')
    imagenponderada = cv2.cvtColor(imagenponderada, cv2.COLOR_BGR2RGB)
    plt.imshow(imagenponderada)
    plt.show()

transfLineal('../LogoRM2.jpeg', 0.7 , 100)
```

```python
def transfLog(im, s):
    image = cv2.imread(im)
    ancho = image.shape[0]
    alto = image.shape[1]
    gray = np.zeros((ancho,alto))
    for i in range(0,ancho):
        for j in range(0,alto):
            pixel = image[i,j]
            gray[i,j]=( (0.3 * pixel[2]) + (0.59 * pixel[1]) + (0.11 * pixel[0]) )
            gray[i,j]=s * np.log(1 +gray[i,j] )
    normalized_image = cv2.normalize(gray, None, 0, 255, cv2.NORM_MINMAX, dtype = cv2.CV_8U)
    cv2.imwrite('binary.jpg', normalized_image) 
    
    imagenponderada = cv2.imread('binary.jpg')
    imagenponderada = cv2.cvtColor(imagenponderada, cv2.COLOR_BGR2RGB)
    plt.imshow(imagenponderada)
    plt.show()

transfLog('../LogoRM2.jpeg', 40)
```

```python
def transfExp(im, s):
    image = cv2.imread(im)
    ancho = image.shape[0]
    alto = image.shape[1]
    gray = np.zeros((ancho,alto))
    for i in range(0,ancho):
        for j in range(0,alto):
            pixel = image[i,j]
            gray[i,j]=( (0.3 * pixel[2]) + (0.59 * pixel[1]) + (0.11 * pixel[0]) )
            gray[i,j]=s * pow( gray[i,j], s )
    normalized_image = cv2.normalize(gray, None, 0, 255, cv2.NORM_MINMAX, dtype = cv2.CV_8U)
    cv2.imwrite('binary.jpg', normalized_image) 
    
    imagenponderada = cv2.imread('binary.jpg')
    imagenponderada = cv2.cvtColor(imagenponderada, cv2.COLOR_BGR2RGB)
    plt.imshow(imagenponderada)
    plt.show()

transfExp('../LogoRM2.jpeg', 4)
```
