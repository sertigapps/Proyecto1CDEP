```python
import matplotlib.pyplot as plt
import cv2
import os 
directory = r'C:\Users\Rajnish\Desktop\GeeksforGeeks'
```

```python
def mostrarCanalImagen(imagen,canal):
    print('Imagen Original')
    image = cv2.imread(imagen)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.axis("off")
    plt.imshow(image)
    plt.show()
    copia = image.copy()
    if canal == 1:
        print('Imagen con canal azul')
        copia[:, :, 0] = 0
        copia[:, :, 1] = 0
    if canal == 2:
        print('Imagen con canal verde')
        copia[:, :, 0] = 0
        copia[:, :, 2] = 0
    if canal == 3:
        print('Imagen con canal rojo')
        copia[:, :, 1] = 0
        copia[:, :, 2] = 0
    if canal == 10:
        print('Imagen con canal rojo y verde')
        copia[:, :, 2] = 0
    if canal == 20:
        print('Imagen con canal verde y azul')
        copia[:, :, 0] = 0
    if canal == 30:
        print('Imagen con canal rojo y azul')
        copia[:, :, 1] = 0
    plt.axis("off")
    plt.imshow(copia)
    plt.show()
mostrarCanalImagen('../imagen1.jpg',1)
mostrarCanalImagen('../imagen1.jpg',2)
mostrarCanalImagen('../imagen1.jpg',3)
mostrarCanalImagen('../imagen1.jpg',10)
mostrarCanalImagen('../imagen1.jpg',20)
mostrarCanalImagen('../imagen1.jpg',30)
```
