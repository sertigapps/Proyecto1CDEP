## Implementando Hard Clustering con K-Means 

```python
def crearClustersKmeans(k):
    X = dataF.iloc[:, [0, 1]].values
    m=X.shape[0] #cantidad de ejemplos
    n=X.shape[1] #cantidad de atributos
    centros=np.array([]).reshape(2,0) 
    resultado = {}
    for i in range(k):
        rand = random.randint(0,m-1)
        centros = np.c_[centros,X[rand]]
        resultado[i+1]=[]
    while True:
        distancia = np.array([]).reshape(m,0)
        for K in range(k):
            distTemp = np.sum((X-centros[:,K])**2,axis=1)
            distancia = np.c_[distancia,distTemp]
        c=np.argmin(distancia, axis=1)+1
        temp={}
        for K in range(k):
            temp[K+1]=np.array([]).reshape(2,0)
        for i in range(m):
            temp[c[i]]=np.c_[temp[c[i]],X[i]]

        for K in range(k):
            temp[K+1]=temp[K+1].T

        for K in range(k):
             centros[:,K]=np.mean(temp[K+1],axis=0)
        cambio = False
        for K in range(k):
            if len(temp[K+1]) != len(resultado[K+1]):
                cambio = True
        if cambio :
            resultado = temp
        else:
            break
    color=['red','blue','green','cyan','magenta', 'brown']
    labels=['cluster1','cluster2','cluster3','cluster4','cluster5','cluster6']
    for K in range(k):
        plt.scatter(resultado[K+1][:,0],resultado[K+1][:,1],c=color[K],label=labels[K])
    plt.scatter(centros[0,:],centros[1,:],s=300,c='yellow',label='centros')
    plt.xlabel('Estatura')
    plt.ylabel('Eddad')
    plt.legend()
    plt.show()
    return centros
```

## Implementando Sof Clustering con GMM usando sklearn

```python
def crearGMM(k):
    color=['red','blue','green','cyan','magenta', 'brown','black','gray']
    labels=['cluster1','cluster2','cluster3','cluster4','cluster5','cluster6']
    gmm = GaussianMixture(n_components=k)
    X = dataF.iloc[:, [0, 1]].values
    gmm.fit(X)
    res=gmm.predict(dataF)
    for i in range(len(res)):
        plt.scatter(dataF.iloc[i,0],dataF.iloc[i,1],c=color[res[i]])
    plt.xlabel('Estatura')
    plt.ylabel('Eddad')
    plt.legend()
    plt.show()
    return gmm
```
