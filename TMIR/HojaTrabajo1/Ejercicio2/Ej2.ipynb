```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
def recrearDe3D(imagen):
    print('Imagenes Original')
    imagev = cv2.imread('../'+imagen+'/'+imagen+'_salida_gray_verde.jpg',0)
    imageor = cv2.cvtColor(imagev, cv2.COLOR_GRAY2RGB)
    plt.axis("off")
    plt.imshow(imageor)
    plt.show()
    imager = cv2.imread('../'+imagen+'/'+imagen+'_salida_gray_rojo.jpg',0)
    imageorr = cv2.cvtColor(imager, cv2.COLOR_GRAY2RGB)
    plt.axis("off")
    plt.imshow(imageorr)
    plt.show()
    
    imagea = cv2.imread('../'+imagen+'/'+imagen+'_salida_gray_azul.jpg',0)
    imageora = cv2.cvtColor(imagea, cv2.COLOR_GRAY2RGB)
    plt.axis("off")
    plt.imshow(imageora)
    plt.show()
    
    blank_image = np.zeros((imagea.shape[0],imagea.shape[1],3), np.uint8)
    blank_image[:, :, 0] = imager
    blank_image[:, :, 1] = imagev
    blank_image[:, :, 2] = imagea
    plt.imshow(blank_image)
    plt.show()
recrearDe3D('perro')
recrearDe3D('imagen1')
recrearDe3D('imagen2')
```
