# CNN para identificacion de lenguaje por medio de audio

# Resumen Paper
https://arxiv.org/pdf/1905.04348.pdf

# Multiclass Language Identification using Deep Learning on Spectral Images of Audio Signals

Recientemente, los asistentes de voz se han convertido en un elemento básico en los productos estrellas de muchas
grandes empresas de tecnología como Google, Apple, Amazon y Microsoft. El desafío para los productos de asistente de voz es que el idioma que usa un usuario necesita ser preestablecido.
La técnica descrita en este documento, identificación de idioma para espectrogramas de audio (LIFAS), utiliza espectrogramas de señales de audio sin procesar como entrada a una red neuronal convolucional (CNN) que se utilizará para la identificación de idiomas. Un beneficio de este proceso es que requiere un preprocesamiento mínimo. De hecho, solo la data en crudo de las señales de audio se introducen en la red neuronal y los espectrogramas se generan cuando cada lote se introduce en la red durante el entrenamiento. Otro beneficio es que la técnica puede utilizar segmentos de audio cortos (aproximadamente 4 segundos) para una clasificación eficaz, necesaria para los asistentes de voz que necesitan identificar idioma tan pronto como un hablante comienza a hablar.

En este paper se utilizo una combinacion de una red residual y una convolucional , en la cual los espectogramas se generaban en tiempo real y no se guardaban en memoria, utilizaron imagenes mas grandes de las que yo utilize para mi red ya que estoy experimentando aun, en el paper explican como por medio del espectograma se pueden extraer caracteristicas visuales del audio las cuales pueden ayudar a reconocer el lenguaje o encontrar patrones, yo utilizare un dataset de mi ambiente laboral y probablemente necesite mas tiempo para lograr los resultados requeridos ya que los archivos de audio no son estandar como los que utilzaron en este paper, pero de cualquier forma ellos lograron un accuracy de 96%

```python

LENGUAJES_DICT = {'en':0,'fr':1,'es':2}
LABELS =  list(LENGUAJES_DICT.keys())
modeloMLP = keras.models.load_model('modeloMLP.h5')
sc=load('std_scaler.bin')
def removertags(text):
    return re.sub(r'<[^<]+?>', '', text)

def removerparentesis(text):
    return re.sub("[\(\[].*?[\)\]]", "", text)

def removerlineas(text):
    return text.replace('\n', ' ') 
    
def removerespacios(text):
    return re.sub(r'\s+', ' ', text)

def limpiar_text(text):
    text = removertags(text)
    text = removerlineas(text)
    text = removerlineas(text)
    text = removerespacios(text)
    return text
def obtener_sample(file_content,start_index,sample_size):
    while not (file_content[start_index].isspace()):
        start_index += 1
    while file_content[start_index].isspace():
        start_index += 1
    end_index = start_index+sample_size 
    while not (file_content[end_index].isspace()):
        end_index -= 1
    return file_content[start_index:end_index]
def definir_alfabeto():
    base_en = 'abcdefghijklmnopqrstuvwxyz'
    special_chars = ' !?¿¡'
    french = 'àâæçéèêêîïôœùûüÿ'
    spanish = 'áéíóúüñ'
    all_lang_chars = base_en + french + spanish 
    small_chars = list(set(list(all_lang_chars)))
    small_chars.sort() 
    big_chars = list(set(list(all_lang_chars.upper())))
    big_chars.sort()
    small_chars += special_chars
    letters_string = ''
    letters = small_chars + big_chars
    for letter in letters:
        letters_string += letter
    return small_chars,big_chars,letters_string
alfabeto = definir_alfabeto()
def contar_chars(text, alphabet):
    alphabet_counts = []
    for letter in alphabet:
        count = text.count(letter)
        alphabet_counts.append(count)
    return alphabet_counts

def obtener_fila(content,start_index,sample_size, alphabet):
    sample_text = obtener_sample(content,start_index,sample_size)
    counted_chars_all = contar_chars(sample_text.lower(), alphabet[0])
    counted_chars_big = contar_chars(sample_text, alphabet[1])
    all_parts = counted_chars_all + counted_chars_big
    return all_parts
def predecir(TEXT):
    cleaned_text = limpiar_text(TEXT)
    while(len(cleaned_text) < 150 ):
        cleaned_text += ' ' + cleaned_text
    input_row = obtener_fila(cleaned_text, 0, 140, alfabeto)
    test_array = sc.transform([input_row])
    raw_score = modeloMLP.predict(test_array)
    pred_idx= np.argmax(raw_score, axis=1)[0]
    score = raw_score[0][pred_idx]*100
    prediction = LABELS[modeloMLP.predict_classes(test_array)[0]]
    return prediction
def remove_html_tags(text):
    """Remove html tags from a string"""
    import re
    clean = re.compile('<.*?>')
    text = re.sub(clean, ' ', text)
    regex = re.compile("\((.*?)\)")
    text = re.sub(regex,'', text)
    regex = re.compile("&.*?;")
    text = re.sub(regex,'', text)
    return text
    
```

# Espectograma 
Vamos a convertir todos los audios a un espectograma para poder analizarlos para lograr una forma estandar vamos a usar]
ffmpeg para normarlizar y cortarlo a los ultimos 5 segundos del audio

```python
def audio_to_spectogram(path, height=192, width=192):
    try:
        signal, sr = librosa.load(path) #this will cut away 5% from start and end
        hl = signal.shape[0]//(width)
        spec = librosa.feature.melspectrogram(signal, n_mels=height, hop_length=int(hl))
        img = librosa.power_to_db(spec)
        start = (img.shape[1] - width) // 2
        return img[:, start:start+width]
    except:
        return []
```
