```python
import matplotlib.pyplot as plt
import cv2
import statistics
import numpy as np
```

```python
def mostrarImagenes(img,p1,p2,p3,p4,p5):
    img = cv2.imread(img)
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    edges = cv2.Canny(gray,p1,p2,apertureSize = p3)

    lines = cv2.HoughLines(edges,1,np.pi/p4,p5)
    for i in range(len(lines)):
        for rho,theta in lines[i]:
            a = np.cos(theta)
            b = np.sin(theta)
            x0 = a*rho
            y0 = b*rho
            x1 = int(x0 + 1000*(-b))
            y1 = int(y0 + 1000*(a))
            x2 = int(x0 - 1000*(-b))
            y2 = int(y0 - 1000*(a))

            cv2.line(img,(x1,y1),(x2,y2),(0,0,255),2)

    plt.imshow(img,cmap = 'gray')
    plt.title('Lineas'), plt.xticks([]), plt.yticks([])
    
    plt.show()
mostrarImagenes('../campos/campo1.jpg',50,150,3,6000,200)
mostrarImagenes('../campos/campo2.jpg',50,150,3,1000,130)
mostrarImagenes('../campos/campo3.jpg',50,150,7,2000,150)
```
